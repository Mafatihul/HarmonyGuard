```python
# uji coba penghapusan non alpha numeric

rx = "x[A-Za-z0-9]+" # hapus pola x70 xaa xf0 x9f dan seterusnya
text = "ya bani taplak kawan kawan xf0 x9f x98 x84 xf0 x9f x98 x84 xf0 x9f x98 x84"
print( re.sub(rx, '', text) )
```

```python
# fungsi penghapusan alpha numeric
def remove_nonaplhanumeric(text):
    text = re.sub('x[A-Za-z0-9./]+', '', text)
    return text
```

```python
# excute penghpausan alpha numeric
df['Tweet'] = df['Tweet'].astype(str)
df ['Tweet'] = df ['Tweet'].apply(remove_nonaplhanumeric)
df
```

```python
# hapus white space 1x 2x 3x dan seterusnya
def remove_space (text):
    text = re.sub(r'\s+', ' ', text)
    return text

text_space = "This     is    an   example    with                                             multiple spaces."
try_space = remove_space(text_space)
print (try_space)
print (text_space)
```

```python
# excute penghpausan white space
df ['Tweet'] = df ['Tweet'].apply(remove_space)
df
```
